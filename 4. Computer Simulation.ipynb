<font size=5;> $$\mu_M:=\frac{1}{M}\sum^M_{i=1} \xi_i \tag{4.1}$$</font>

```python
def kernel_density_estimate(observations):
    a={}
    for obs in observations:
        bucket=(int)(obs/0.05)
        #print(n)
        try:
            a[bucket]+=1
        except KeyError:
            a[bucket]=1
    return a
nrows=4
ncols=1
figsize=8
fig,axes=plt.subplots(nrows=nrows, ncols=ncols, figsize=(figsize,nrows*figsize))
M=1000
for i in range(nrows):
    m_i=M*(10**i)
    delta_x=0.05
    observations=np.random.uniform(size=m_i)
    a=kernel_density_estimate(observations)
    keys=sorted(a)
    axes[i].plot(keys, [a[key]/(m_i*delta_x) for key in keys], '.-')
    axes[i].set_ylim([-0.05, 1.5])
```

Here, we generated sets of $U(0, 1)$ samples $\{\xi_i \}^{n}_{i=1}$, with $n = 10^3$. These were combined to give samples of the form

<font size=4>$$\frac{\sum_{i=1}^{n} \xi_i -n\mu}{\sigma \sqrt{n}} \tag{4.7}$$</font>

where $\mu = 1/2$ and $\sigma^2 = 1/12$. We repeated this $M = 10^4$ times. These $M$ data points were then used to obtain a kernel density estimate.

```python
M=1000000
n=1000
transformed_samples=[]
for i in range(M):
    uniform_samples=np.random.uniform(size=n)
    transformed_samples.append((np.sum(uniform_samples)-n*0.5)/(np.sqrt(n/4)))
# print(transformed_samples[:5])
fig, axes=plt.subplots(nrows=1, ncols=1, figsize=(10,10))
a=kernel_density_estimate(transformed_samples)
keys=sorted(a)
axes.plot(keys, [a[key]/(m_i*delta_x) for key in keys], '.-')
# axes[i].set_ylim([-0.05, 1.5])
```
